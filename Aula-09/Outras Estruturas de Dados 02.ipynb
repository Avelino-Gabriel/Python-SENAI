# Range

### Estrutura:

- range(tamanho)

ou 

- range(inicio, fim)

ou

- range(inicio, fim, passo)

```python
#uso mais comum no for:
produtos = ['arroz', 'feijao', 'macarrao', 'atum', 'azeite']
estoque = [50, 100, 20, 5, 80]

for c in range(0, len(produtos)):
    print(produtos[c])

for i in range(0, len(estoque)):
    print(estoque[i])
```

```python
#range com inicio e fim
for c in range(1, 10):
    print(c)

#vamos olhar no for para entender
```

- Exemplo: Modelo Jack Welch da G&E

    1. Classe A: 10% melhor
    2. Classe B: 80% mantém/busca melhorar
    3. Classe C: 10% pior
    
Quem são os funcionários classe B?

```python
funcionarios = ['Maria', 'José', 'Antônio', 'João', 'Francisco', 'Ana', 'Luiz', 'Paulo', 'Carlos', 'Manoel', 'Pedro', 'Francisca', 'Marcos', 'Raimundo', 'Sebastião', 'Antônia', 'Marcelo', 'Jorge', 'Márcia', 'Geraldo']
vendas = [2750, 1900, 1500, 1200, 1111, 1100, 999, 900, 880, 870, 800, 800, 450, 400, 300, 300, 120, 90, 80, 70]
classe_A = []
valores_classe_A = []
classe_B = []
valores_classe_B = []
classe_C = []
valores_classe_C = []

porcentagem = len(vendas) * (10/100)
for c in range(0, int(porcentagem)):
    minimo = min(vendas)
    maximo = max(vendas)
    nome_maximo = funcionarios[vendas.index(maximo)]
    nome_minimo = funcionarios[vendas.index(minimo)]
    valores_classe_A.append(maximo)
    valores_classe_C.append(minimo)
    classe_A.append(nome_maximo)
    classe_C.append(nome_minimo)
    vendas.remove(minimo)
    vendas.remove(maximo)
    funcionarios.remove(f'{nome_maximo}')
    funcionarios.remove(f'{nome_minimo}')

valores_classe_B = vendas
classe_B = funcionarios

while True:
    try:
        opcao = int(input('''[1] Exibir Classe A
[2] Exibir Classe B
[3] Exibir Classe C
[4] Encerrar programa'''))
    except: 
        print('Digite uma opção válida')
    else:
        if opcao == 1:
            for a in range (0, len(classe_A)):
                print(f'{classe_A[a]}: {valores_classe_A[a]}')
            print('-' * 30)
        elif opcao == 2:
            for b in range (0, len(classe_B)):
                print(f'{classe_B[b]}: {valores_classe_B[b]}')
            print('-' * 30)
        elif opcao == 3:
            for c in range (0, len(classe_C)):
                print(f'{classe_C[c]}: {valores_classe_C[c]}')
            print('-' * 30)
        elif opcao == 4:
            print('SAINDO DO PROGRAMA...')
            break
        else:
            print('Digite uma opção válida!')
print('PROGRAMA ENCERRADO! Volte Sempre!')
```
