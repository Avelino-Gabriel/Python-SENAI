#### Exercício 1:
Crie um programa que liste todos os arquivos em um diretório específico usando os.listdir()

```python
import os
from pathlib import Path

print('-' * 50)
print('LISTA DE ARQUIVOS'.center(50))
print('-' * 50)
for a in os.listdir(Path.cwd()):
    if os.path.isfile(os.path.join(Path.cwd(), a)):
        print(a)
```

#### Exercício 2:
Crie um programa que conte quantos arquivos há em um diretório específico usando os.listdir()


```python
import os
from pathlib import Path

contador = 0

print('-' * 30)
print('QUANTIDADE DE ARQUIVOS'.center(30))
print('-' * 30)
for a in os.listdir(Path.cwd()):
    if os.path.isfile(os.path.join(Path.cwd(), a)):
        contador += 1
print(f'Quantidade de arquivos = {contador}')
```

#### Exercício 3:
Escreva um programa que crie um novo diretório usando os.mkdir()

```python
import os
while True:
    try:
        nome_diretório = str(input('Nome do Diretório: ')).strip()
        if nome_diretório == '':
            break
        else:
            os.mkdir(nome_diretório)
            print(f'Diretório {nome_diretório} criado com sucesso!')
            break
    except:
        print(f'O diretório {nome_diretório} já existe!')
```

#### Exercício 4:
Crie um programa que verifique se um arquivo existe em um diretório usando os.path.exists()

```python
import os
while True:
    pesquisar_diretório = str(input('Nome do Diretório: ')).strip()
    if pesquisar_diretório == '':
        break
    if os.path.exists(pesquisar_diretório):
        print(f'Diretório {pesquisar_diretório} foi encontrado!')
    else:
        print(f'Diretório {pesquisar_diretório} não existe!')
```

#### Exercício 5:
Crie um programa que leia o conteúdo de um arquivo e o imprima na tela usando open()

```python
from pathlib import Path
while True:
    pesquisar_arquivo = str(input('Nome do Arquivo: ')).strip()
    caminho_arquivo = Path.cwd() / 'arquivos' / pesquisar_arquivo
    if caminho_arquivo.exists() and pesquisar_arquivo != '':
        with open(caminho_arquivo, 'r') as arquivo:
            text = arquivo.readlines()
            print(f"Conteúdo do arquivo {pesquisar_arquivo}:")
            for linha in text:
                print(linha, end='')
        break
    elif pesquisar_arquivo == '':
        break
    else:
        print(f'Arquivo {pesquisar_arquivo} não existe!')
```

#### Exercício 6:
Crie um programa que escreva uma lista de strings em um arquivo de texto usando open()

```python
from pathlib import Path

nome_arquivo = str(input('Nome do Arquivo: ')).strip()
caminho_arquivo = Path.cwd() / 'arquivos' / f'{nome_arquivo}.txt'
text = []
while True:
    write = str(input('Escreva uma linha do seu texto: ')).strip()
    if write == '':
        break
    else:
        text.append(write)

with open(caminho_arquivo, 'w', encoding='utf8') as arquivo:
    for linha in text:
        arquivo.write(F'{linha}\n')

print(f'Arquivo {nome_arquivo} criado!')
```

#### Exercício 7:
Crie um programa que copie um arquivo de um diretório para outro usando shutil.copy()

```python
from pathlib import Path
import shutil

caminho_arquivo = Path.cwd()
nome_arquivo = '' 
while True:
    try:
        nome_caminho = str(input('Diretório do Arquivo: ')).strip()
        if nome_caminho == '':
            break
        nome_arquivo = str(input('Nome do Arquivo: ')).strip()
        if nome_arquivo == '':
            break
        caminho_copia = str(input('Diretório para criar cópia: ')).strip()
        if caminho_copia == '':
            break
        shutil.copy(Path(caminho_arquivo, nome_caminho, nome_arquivo), Path(caminho_arquivo, caminho_copia))
        break
    except:
        print('Erro ao fazer a cópia, componente não detectado!')    

print(f'Arquivo "{nome_arquivo}" copiado!')
```

#### Exercício 8:
Crie um programa que renomeie um arquivo usando os.rename()

```python
from pathlib import Path
import os

nome_arquivo = ''
novo_nome = ''
caminho_arquivo = Path.cwd() 
while True:
    try:
        nome_caminho = str(input('Nome do Diretório: ')).strip()
        if Path(caminho_arquivo, nome_caminho).exists() and nome_caminho != '':
            nome_arquivo = str(input('Nome do Arquivo: ')).strip()
            if Path(caminho_arquivo, nome_caminho, nome_arquivo).exists() and nome_arquivo != '':
                novo_nome = str(input('Novo Nome do Arquivo: ')).strip()
                os.rename(Path(caminho_arquivo, nome_caminho, nome_arquivo), Path(caminho_arquivo, nome_caminho, novo_nome))
                break
            elif nome_arquivo == '':
                break
            else:
                print(f'Arquivo {nome_arquivo} não existe!')
        elif nome_caminho == '':
            break
        else:
                print(f'Diretório {nome_caminho} não existe!')
    except:
        print('Erro ao renomear arquivo!')    

print(f'Arquivo "{nome_arquivo}" renomeado para "{novo_nome}"!')
```

#### Exercício 9:
Crie um programa que exclua um arquivo usando os.remove()

```python
from pathlib import Path
import os

nome_arquivo = ''
caminho_arquivo = Path.cwd() 
while True:
    try:
        nome_caminho = str(input('Nome do Diretório: ')).strip()
        if Path(caminho_arquivo, nome_caminho).exists() and nome_caminho != '':
            nome_arquivo = str(input('Nome do Arquivo: ')).strip()
            if Path(caminho_arquivo, nome_caminho, nome_arquivo).exists() and nome_arquivo != '':
                os.remove(Path(caminho_arquivo, nome_caminho, nome_arquivo))
                break
            elif nome_arquivo == '':
                break
            else:
                print(f'Arquivo {nome_arquivo} não existe!')
        elif nome_caminho == '':
            break
        else:
                print(f'Diretório {nome_caminho} não existe!')
    except:
        print('Erro ao remover arquivo!')    

print(f'Arquivo "{nome_arquivo}" removido!')
```

#### Exercício 11:
Crie um programa que liste todos os arquivos em um diretório específico usando a biblioteca pathlib com o método iterdir()

```python
from pathlib import Path

caminho_arquivo = Path.cwd() 

nome_diretório = str(input('Nome do Diretório: ')).strip()

print('-' * 100)
print('LISTA DE DIRETÓRIOS'.center(100))
print('-' * 100)
for v in Path.iterdir(Path(caminho_arquivo, nome_diretório)):
    if v.is_file():
        print(v)
print('-' * 100)
```
