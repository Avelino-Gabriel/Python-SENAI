```python
# codigo aqui
from math import floor


def title(text):
    texto = f'''{'-' * 138}
{text.center(138)}
{'-' * 138}'''
    return texto


def menu(opcoes):
    while True:
        print('-'*138)
        try:
            opcao = int(input(opcoes))
            print('-'*138)
            return opcao
        except:
            print('-'*138)
            print('Opção Inválida')
            print('-'*138)


def avaliacao(nota):
    while True:
        try:
            avaliar = int(input(nota))
            if avaliar > 5 or avaliar < 1:
                print('Nota Inválida!')
            else:
                return avaliar
        except:
            print('Nota Inválida!')


def cadastro(student, book, date1, date2):
    aluno = str(input(student)).strip().title()
    titulo = str(input(book)).strip().title()
    data_emprestimo = str(input(date1)).strip().title()
    data_devolucao = str(input(date2)).strip().title()

    return (aluno, titulo, data_emprestimo, data_devolucao)


biblioteca = {}
livro_notas = {}
notas_atendente = []
biblioteca['livros'] = {}

while True:
    print(title('BIBLIOTECA'))
    livro = {}

    opcao_menu_principal = menu('''[1] Mostrar Relatório
[2] Cadastrar empréstimo
[3] Indicar devolução
[4] Dar avaliação ao livro
[5] Dar avaliação ao atendimento
[6] Sair
Sua opção:''')
    if opcao_menu_principal == 1:
        while True:
            opcao_relatorio = menu('''[1] Mostrar Livros
[2] Mostrar nota de atendimento
[3] Mostrar relatório completo
[4] sair
Sua opção:''')
            if opcao_relatorio == 1:
                while True:
                    opcao_relatorio_livros = menu('''[1] Mostrar Livro Específico
[2] Mostrar todos os livros
[3] sair
Sua opção:''')
                            
                    if opcao_relatorio_livros == 1:
                        titulo = str(input('Título do livro: ')).strip().title()
                        if titulo == '':
                            break
                        try:
                            print(title(titulo))
                            for v in biblioteca['livros'][titulo].items():
                                if 'empréstimos' == v[0]:
                                    print(f'{v[0]}: ')
                                    for a in v[1]:
                                        print(a)
                                    print()
                                else:
                                    print(f'{v[0]}: {v[1]}')
                            print('-' * 138)
                        except:
                            print('Livro não encontrado')
                            pass
                    elif opcao_relatorio_livros == 2:
                        if biblioteca['livros'] == {}:
                            print('Não há livros cadastrados no momento')
                        else:
                            for l in biblioteca['livros'].items():
                                print(title(l[0]))

                                for c, v in enumerate(l):
                                    if c == 0:
                                        pass
                                    else:
                                        for a in v.items():
                                            if 'empréstimos' == a[0]:
                                                print(f'{a[0]}: ')
                                                for e in a[1]:
                                                    print(e)
                                                print()
                                            else:
                                                print(f'{a[0]}: {a[1]}')
                                print('-' * 138)
                    elif opcao_relatorio_livros == 3:
                        break
                    else:
                        print('Opção Inválida')
            elif opcao_relatorio == 2:
                try:
                    print('-' * 138)
                    print(f'Nota de Atendimento: {biblioteca["nota de atendimento"]}')
                    print('-' * 138)
                except:
                    print('Não há uma nota de atendimento no momento')
                    pass
            elif opcao_relatorio == 3:
                for b in biblioteca.items():
                    for x, k in enumerate(b):
                        if k == 'livros':
                            print(title(k.upper()))
                        else:
                            if type(k) is dict:
                                for l in k.items():
                                    print(title(l[0]))
                                    for c, v in enumerate(l):
                                        if c == 0:
                                            pass
                                        else:
                                            for a in v.items():
                                                if 'empréstimos' == a[0]:
                                                    print(f'{a[0]}: ')
                                                    for e in a[1]:
                                                        print(e)
                                                    print()
                                                else:
                                                    print(f'{a[0]}: {a[1]}')
                            else:
                                print('-' * 138)
                                if x % 2 == 0:
                                    print(f'{k}: ', end='')
                                else:
                                    print(k)
                    print('-' * 138)  
            elif opcao_relatorio == 4:
                break
            else:
                print('Opção Inválida')

    elif opcao_menu_principal == 2:
        aluno_emprestar, titulo_emprestar, data_emprestimo_emprestar, data_devolucao_emprestar = cadastro('Nome: ', 'Título do Livro: ', 'Data de Empréstimo: ', 'Data de Devolução: ')
        
        if aluno_emprestar == '' or titulo_emprestar == '' or data_emprestimo_emprestar == '' or data_devolucao_emprestar == '':
            break

        if titulo_emprestar not in biblioteca['livros']:
            emprestimos = []
            ficha = {'nome': aluno_emprestar, 'data de empréstimo': data_emprestimo_emprestar, 'data de devolução': data_devolucao_emprestar, 'status de devolução': 'PENDENTE'}
            emprestimos.append(ficha.copy())
            livro['empréstimos'] = emprestimos.copy()
            biblioteca['livros'][titulo_emprestar] = livro.copy()
        else:
            ficha = {'nome': aluno_emprestar, 'data de empréstimo': data_emprestimo_emprestar, 'data de devolução': data_devolucao_emprestar, 'status de devolução': 'PENDENTE'}
            if 'empréstimos' in biblioteca['livros'][titulo_emprestar]:
                biblioteca['livros'][titulo_emprestar]['empréstimos'].append(ficha.copy())
            else:
                emprestimos = []
                emprestimos.append(ficha.copy())
                livro['empréstimos'] = emprestimos.copy()
                biblioteca['livros'][titulo_emprestar] = livro.copy()
        print('-' * 138)
        print('Empréstimo adicionado')

    elif opcao_menu_principal == 3:
        aluno_devolver, titulo_devolver, data_emprestimo_devolver, data_devolucao_devolver = cadastro('Nome: ', 'Título do Livro: ', 'Data de Empréstimo: ', 'Data de Devolução: ')
        
        if aluno_devolver == '' or titulo_devolver == '' or data_emprestimo_devolver == '' or data_devolucao_devolver == '':
            break

        if titulo_devolver not in biblioteca['livros']:
            emprestimos = []
            ficha = {'nome': aluno_devolver, 'data de empréstimo': data_emprestimo_devolver, 'data de devolução': data_devolucao_devolver, 'status de devolução': 'DEVOLVIDO'}
            emprestimos.append(ficha.copy())
            livro['empréstimos'] = emprestimos.copy()
            biblioteca['livros'][titulo_devolver] = livro.copy()
        else:
            ficha = {'nome': aluno_devolver, 'data de empréstimo': data_emprestimo_devolver, 'data de devolução': data_devolucao_devolver, 'status de devolução': 'DEVOLVIDO'}
            if 'empréstimos' in biblioteca['livros'][titulo_devolver]:
                status_pesquisa = False
                indice = 0
                for f in biblioteca['livros'][titulo_devolver]['empréstimos']:
                    if ficha['nome'] == f['nome'] and ficha['data de empréstimo'] == f['data de empréstimo']:
                        status_pesquisa = True
                        indice = biblioteca['livros'][titulo_devolver]['empréstimos'].index(f)
                    else: 
                        status_pesquisa = False
                if status_pesquisa == True:
                    biblioteca['livros'][titulo_devolver]['empréstimos'][indice] = ficha.copy()
                else:
                    biblioteca['livros'][titulo_devolver]['empréstimos'].append(ficha.copy())
            else:
                emprestimos = []
                emprestimos.append(ficha.copy())
                biblioteca['livros'][titulo_devolver]['empréstimos'] = emprestimos.copy()
            print('-' * 138)
            print('Informações de devolução adicionadas')

    elif opcao_menu_principal == 4:
        nome = str(input('Nome do Aluno: ')).strip().title()
        titulo = str(input('Título do Livro: ')).strip().title()
        if nome == '':
            break

        if titulo == '':
            break
        
        for l in biblioteca['livros'].values():
            for emprestimo in l['empréstimos']:
                for chave,valor in emprestimo.items():
                    if (emprestimo.get("nome") == nome):
                        avaliacao_livro = avaliacao('Digite uma nota para o livro de 1 a 5 (999 para cancelar operação): ')
                        if avaliacao_livro == 999:
                            break
                                
                        if titulo not in biblioteca['livros']:
                            notas = []
                            notas.append(avaliacao_livro)
                            livro_notas[titulo] = notas.copy()
                            media_livro = sum(livro_notas[titulo]) / len(livro_notas[titulo])
                            livro['média do livro'] = '*' * floor(int(media_livro))
                            biblioteca['livros'][titulo] = livro.copy()
                        else:
                            if 'média do livro' in biblioteca['livros'][titulo]:
                                livro_notas[titulo].append(avaliacao_livro)
                                media_livro = sum(livro_notas[titulo]) / len(livro_notas[titulo])
                                biblioteca['livros'][titulo]['média do livro'] = '*' * floor(int(media_livro))
                            else:
                                notas = []
                                notas.append(avaliacao_livro)
                                livro_notas[titulo] = notas.copy()
                                media_livro = sum(livro_notas[titulo]) / len(livro_notas[titulo])
                                biblioteca['livros'][titulo]['média do livro'] = '*' * floor(int(media_livro))
                        print('-' * 138)
                        print('Nota do livro adicionada')
                        break
                    else:
                        print("livro não emprestado para esse aluno, avaliação cancelada")
                        break

    elif opcao_menu_principal == 5:
        avaliacao_atendimento = avaliacao('Digite uma nota para o atendente de 1 a 5 (999 para cancelar operação): ')

        if avaliacao_atendimento == 999:
            break

        notas_atendente.append(avaliacao_atendimento)
        media_atendimento = sum(notas_atendente) / len(notas_atendente)
        biblioteca['nota de atendimento'] = '*' * floor(int(media_atendimento))
        print('-' * 138)
        print('Nota de atendimento adicionada')
        
    elif opcao_menu_principal == 6:
        break
    else:
        print('Opção Inválida!')
print('PROGRAMA FINALIZADO! Volte Sempre!')

    

```
