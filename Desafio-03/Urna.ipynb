```python
def menu(opc):
    while True:
        try:
            opcao = int(input(opc))
            if opcao > 0 and opcao < 5:
                return opcao
        except: 
            print('Digite uma opção válida!')


def cadastrar_candidato(nome, num, partido):
    cand = {
        'nome': nome,
        'número': num,
        'partido': partido,
        'quantidade de votos': 0
    }

    return cand


def iniciarVotacao():
    while True:
        voto = str(input('Número do Candidato: ')).strip()
        if voto.isnumeric():
            return voto
        elif voto == '':
            return voto
        else:
            print('Input inválido!')


def voto_branco():
    cand = {
        'nome': 'branco',
        'número': '',
        'partido': '-',
        'quantidade de votos': 1
    }

    return cand


def voto_nulo(num):
    cand = {
        'nome': 'nulo',
        'número': num,
        'partido': '-',
        'quantidade de votos': 1
    }

    return cand

def relatorio(arr):
    for a in arr.values():
        if type(a) is list:
            print('-'*138)
            print(f'{"Nome":<25}{"Número":<25}{"Partido":<25}{"Quantidade de Votos":<25}{"Porcentagem de Votos":<25}')
            print('-'*138)
            for c in a:
                for v in c.values():
                    if 'porcentagem' in c and v == c['porcentagem']:
                        print(f'{v:<25.2%}', end='')
                    else:   
                        print(f'{v:<25}', end='')
                print()
            print('-'*138)
        else:
            print(f'Total de Votos: {a}')
            print('-'*138)


def calculo_porcentagem(data):
    for f, p in enumerate(data['candidatos']):
        data['candidatos'][f]['porcentagem'] = data['candidatos'][f]["quantidade de votos"]/ data["total de votos"]


def maior(votacao):
    porcents = []
    maior = 0
    nome_maior = ''
    for u in votacao['candidatos']:
        if u['nome'] != 'nulo' and u['nome'] != 'branco':
            porcents.append((u['nome'], u['porcentagem']))
    for value in porcents:
        if value[1] > maior:
            maior = value[1]
            nome_maior = value[0]
    return(nome_maior, maior)
 


def Votacao():
    p = 0
    votacao = {
        'candidatos': [],
        'total de votos': 0
    }
    while True:
        opc_menu_principal = menu('''[1] Cadastrar candidatos
[2] Iniciar votação
[3] Prévia de votos
[4] Encerrar votação
Sua Opção: ''')
        if opc_menu_principal == 1:
            if p == 0: 
                while True:
                    nome = str(input('Nome do Candidato: ')).strip().title()
                    numero = str(input('Número do candidato: ')).strip()
                    partido = str(input('Partido do candidato: ')).strip().title()
                    if nome == '' or numero == '' or partido == '':
                        break
                    else:
                        candidato = cadastrar_candidato(nome, numero, partido)
                        votacao['candidatos'].append(candidato)
                p += 1
            else:
                print('Não é possível cadastrar mais candidatos!')

        elif opc_menu_principal == 2:
            while True:
                voto = iniciarVotacao()
                c = 0
                ind = 0
                nome = ''
                while c < len(votacao['candidatos']):
                    if voto == votacao['candidatos'][c]['número']:
                        encontrado = True
                        break
                    else:
                        encontrado = False
                        if votacao['candidatos'][c]['nome'] == 'nulo':
                            nome = votacao['candidatos'][c]['nome']
                            ind = c
                        c += 1

                if encontrado == True:
                    votacao['candidatos'][c]['quantidade de votos'] += 1
                    votacao['total de votos'] += 1
                    calculo_porcentagem(votacao)
                    
                elif encontrado == False and voto == '':
                    branco = voto_branco()
                    if len(votacao['candidatos']) < 4 and votacao['candidatos'][-1]['nome'] != 'nulo':
                        votacao['candidatos'].append(branco)
                        
                    elif len(votacao['candidatos']) < 4 and votacao['candidatos'][-1]['nome'] == 'nulo':
                        votacao['candidatos'].insert(len(votacao['candidatos']) - 1, branco)
                    else:
                        votacao['candidatos'].insert(len(votacao['candidatos']) - 2, branco)
                    votacao['total de votos'] += 1
                    calculo_porcentagem(votacao)
                elif encontrado == False and voto == '999999':
                    break
                elif encontrado == False and nome:
                    votacao['candidatos'][ind]['quantidade de votos'] += 1
                    votacao['total de votos'] += 1
                    calculo_porcentagem(votacao)
                else:
                    nulo = voto_nulo(voto)
                    if len(votacao['candidatos']) < 4:
                        votacao['candidatos'].append(nulo)
                    else:
                        votacao['candidatos'].insert(len(votacao['candidatos']) - 2, nulo)
                    votacao['total de votos'] += 1
                    calculo_porcentagem(votacao)
        elif opc_menu_principal == 3:
            relatorio(votacao)

        elif opc_menu_principal == 4:
            relatorio(votacao)
            print(f'{maior(votacao)[0]} venceu a eleição com {maior(votacao)[1]:.2%}')
            break

Votacao()
```
