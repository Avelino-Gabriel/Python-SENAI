```python
def menu(opc):
    while True:
        try:
            opcao = int(input(opc))
            if opcao > 0 and opcao < 5:
                return opcao
        except: 
            print('Digite uma opção válida!')


def cadastrar_candidato(nome, num, partido):
    cand = {
        'nome': nome,
        'número': num,
        'partido': partido,
        'quantidade de votos': 0
    }

    return cand


def iniciarVotacao():
    while True:
        voto = str(input('Número do Candidato: ')).strip()
        if voto.isnumeric():
            return voto
        else:
            print('Input inválido!')


def voto_branco():
    cand = {
        'nome': 'branco',
        'número': '',
        'quantidade de votos': 1
    }

    return cand


def voto_nulo(num):
    cand = {
        'nome': 'nulo',
        'número': num,
        'quantidade de votos': 1
    }

    return cand

def relatorio(arr):
    for c in arr:
        for v in c.values():
            print(f'{v:<10}', end='')
    return


def Votacao():
    votacao = {
        'candidatos': [],
        'total de votos': 0
    }
    while True:
        opc_menu_principal = menu('''[1] Cadastrar candidatos
[2] Iniciar votação
[3] Prévia de votos
[4] Encerrar votação
Sua Opção: ''')
        if opc_menu_principal == 1:
            nome = str(input('Nome do Candidato: ')).strip().title()
            numero = str(input('Número do candidato: ')).strip()
            partido = str(input('Partido do candidato: ')).strip().title()
            candidato = cadastrar_candidato(nome, numero, partido)
            votacao['candidatos'].append(candidato)

        elif opc_menu_principal == 2:
            while True:
                voto = iniciarVotacao()
                c = 0
                ind = 0
                nome = ''
                while c < len(votacao['candidatos']):
                    if voto == votacao['candidatos'][c]['número']:
                        encontrado = True
                        break
                    else:
                        encontrado = False
                        if votacao['candidatos'][c]['nome'] == 'nulo':
                            nome = votacao['candidatos'][c]['nome']
                            ind = c
                        c += 1

                if encontrado == True:
                    votacao['candidatos'][c]['quantidade de votos'] += 1
                elif encontrado == False and voto == '':
                    branco = voto_branco()
                    votacao['candidatos'].append(branco)
                elif encontrado == False and voto == '999999':
                    break
                elif encontrado == False and nome:
                    votacao['candidatos'][ind]['quantidade de votos'] += 1
                else:
                    nulo = voto_nulo(voto)
                    votacao['candidatos'].append(nulo)
                print(votacao)
        elif opc_menu_principal == 3:
            relatorio(votacao['candidatos'])

        elif opc_menu_principal == 4:
            break

Votacao()
```
